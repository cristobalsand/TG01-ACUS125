En el presente documento, se presenta de forma resumida el funcionamiento y modelamiento matemático de un sistema masa-resorte-amortiguador que se utiliza generalmente para simular sistemas vibratorios de 1 grado de libertad en este caso y que se puede visualizar como en la siguiente figura:

<p style="text-align:center">
	<img src="MRA.png" alt="MDOF" width=30%/></a><br>
    <strong> Figura 1: Sistema masa-resorte-amortiguador de un grado de libertad</strong>
</p>

Donde:

* El resorte **k** representa la rigidez del sistema.

* El amortiguador **c** representa el amortiguamiento.

* **m** es la masa.

* **F** es la fuerza exitadora que para este ejemplo es igual a cero.

* **x** es la variable que representa el desplazamiento de la masa respecto al tiempo.

En este sistema el amortiguador cumple la función de disipación de energía tal como ocurre en la realidad y es por esto es que las vibraciones no son infinitas, ya que en la realidad afecta la fuerza de roce entre los materiales o la energía que se pierde en forma de calor. El encargado de hacer que el sistema vibre es el resorte.

Para modelar matemáticamente este problema, se tiene la ecuación diferencial de segundo orden homogénea en donde actúan las fuerzas que se muestran en el diagrama anterior.

\begin{align*}
m\ddot{x}(t)+c\dot{x}(t)+kx(t)&=0
\label{eq1}\tag{1}
\end{align*}

Donde:
    
* $\dot{x}$ es la razón de cambio del desplazamiento respecto al tiempo (velocidad).

* $\ddot{x}$ es la razón de cambio de la velocidad respecto al tiempo (aceleración).

* El lado derecho de la ecuación es igual a cero debido a que la fuerza exitadora es igual a cero.

Además se definen las condiciones iniciales ($t = 0$) para el desplazamiento y la velocidad:

\begin{align*}
x(0)&=x_{0} & \dot{x}(0) =\dot{x}_{0}
\end{align*}

Para acomodar el desarrollo algebraico se reescribe la ecuación dividiendo todos los términos por ''m'' y reemplazando algunos términos por los siguientes parámetros:

\begin{align*}
\omega_{n}&=\sqrt{\frac{k}{m}} \quad {\text{es la frecuencia angular natural}}\,\left[ \frac{rad}{s}\right]
&
\zeta =\frac{c}{2\,\sqrt{k\,m}} \quad {\text{ es la razón de amortiguamiento (adimensional)}}
\end{align*}

Ahora la ecuación ([1](#mjx-eqn-eq1)) queda escrita de la siguiente manera:

\begin{align*}
\ddot{x}(t)+2\zeta \omega_{n}\,\dot{x}(t)+\omega_{n}^{2}\,x(t)&=0
\label{eq2}\tag{2}
\end{align*}

Y se conjetura una solución general de la ecuación del tipo:

\begin{align*}
x(t)&=Ae^{\alpha t}
\label{eq3}\tag{3}
\end{align*}

En la cual si se reemplaza en la ecuación ([2](#mjx-eqn-eq2)) se obtiene la **ecuación característica** para obtener el valor de $\alpha$:

\begin{align*}
\alpha&= \omega_{0}\big(-\zeta\pm\sqrt{\zeta^{2}-1}\big)
\label{eq4}\tag{4}
\end{align*}

Ahora, es posible analizar 4 casos que resuelven la ecuación característica, dependiendo del valor real e imaginario de la razón de amortiguamiento: 

## 1. Caso NO amortiguado

Cuando el coeficiente de amortiguamiento es nulo ($\zeta = 0$) el valor de $\alpha$ es puramente imaginario, por lo que en la solución general la exponencial compleja no posee un término escalar que vaya disminuyendo la amplitud máxima a través del tiempo y que por lo tanto, implica que el sistema no pierde energía.

La solución general de este caso es del tipo:

\begin{align*}
x(t)&=B_{1}e^{j\omega_{n}t}+B_{2}e^{-j\omega_{n}t}
\end{align*}

Y que considerando las condiciones iniciales queda:

\begin{align*}
x(t)&= x_{0} \cos{\omega_{n}t} + \frac{\dot{x_{0}}}{\omega_{n}} \sin{\omega_{n}t}
\label{eq5}\tag{5}
\end{align*}

## 2. Caso sub-amortiguado

En este caso $0<\zeta<1$ y el sistema vibra pero posee un coeficiente real que escala temporalmente a la función exponencial compleja, que se encarga de decaer la amplitud máxima y la frecuencia de oscilación.

La solución general de este caso está dada por la siguiente expresión:

\begin{align*}
x(t)&=B_{1}e^{(-\zeta\omega_{n}-j\omega_{n}\sqrt{1-\zeta^{2}})t}+B_{2}e^{(-\zeta\omega_{n}+j\omega_{n}\sqrt{1-\zeta^{2}})t}
\end{align*}

Pero que al integrar las condiciones iniciales y asignar nuevos parámetros que están incluidos indirectamente en la ecuación, resulta:

\begin{align*}
x(t)&=Ae^{-\zeta\omega_{n}t}\,{\text{sen}}\big( \omega_{d}t-\phi_{d}\big)
\label{eq6}\tag{6}
\end{align*}

Donde:

\begin{align*}
A&=\sqrt{x_{0}^{2}+\big(\frac{\dot{x_{0}}+\zeta\omega_{n}x_{0}}{\omega_{d}}\big)^{2}} \, \text{es la amplitud}; &  \phi={\text{tan}}^{-1}\bigg( \frac{x_{0}\omega_{d}}{\dot{x}_{0}+\zeta\omega_{n}x_{0}} \bigg) \, \text{es la fase y } \, \omega_{d}=\omega_{n}\sqrt{1-\zeta^{2}} \, \text{es la frecuencia natural amortiguada}
\end{align*}

## 3. Caso críticamente amortiguado

Cuando $\zeta=1$, la ecuación característica posee única solución real ($\alpha = -\omega_{0}$) y que elimina la oscilación al no tener componente complejo en la función exponencial dentro de la solución general. La solución para este caso posee dos términos linealmente independientes tales que:

\begin{align*}
x(t)&=e^{-\omega_{n}\,t}\,(C_{1}+C_{2}\,t)
\end{align*}

Que aplicando condiciones iniciales se expresa como:

\begin{align*}
x(t)&=e^{-\omega_{n}\,t}\,\big[\,x_{0}+\big(\dot{x}_{0}+\omega_{n}x_{0}\big)\cdot t\,\big]
\label{eq7}\tag{7}
\end{align*}

## 4. Caso sobre-amortiguado

Cuando $\zeta>1$, la ecuación característica tiene dos soluciones reales que implica que en la solución general no existe una componente compleja, eliminando así la oscilación del sistema y que además va disminuyendo su energía a través del tiempo. La solución general para este caso es:

\begin{align*}
x(t)&=e^{-\zeta t}\bigg[C_{1}e^{-\omega_{n}\big(\sqrt{\zeta^{2}-1}\big)\,t}+C_{2}e^{\omega_{n}\big(\sqrt{\zeta^{2}-1}\big)\,t}\bigg]
\label{eq8}\tag{8}
\end{align*}

Y que en condiciones iniciales:

\begin{align*}
x(t)&=\frac{e^{-\omega_{n}t}}{2\big(\sqrt{\zeta^{2}-1}\big)}\,\cdot\bigg\{\,\bigg[\frac{\dot{x}_{0}}{\omega_{n}}+x_{0}\,\big(\zeta+\sqrt{\zeta^{2}-1}\,\big)\bigg]\,e^{\omega_{n}\big(\sqrt{\zeta^{2}-1}\big)\,t}\,+\,\bigg[-\frac{\dot{x}_{0}}{\omega_{n}}+x_{0}\,\big(-\zeta+\sqrt{\zeta^{2}-1}\,\big)\bigg]\,e^{-\omega_{n}\big(\sqrt{\zeta^{2}-1}\big)\,t}\,\bigg\}
\label{eq9}\tag{9}
\end{align*}

# Representación gráfica

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline 
```

```python
# Caso NO amortiguado

t = np.linspace(0,10,1000) # vector tiempo

k = 300
m = 30
c = 0

omega_n = np.sqrt(k/m)
zeta = c/(2*np.sqrt(k*m))

x_0 = 0.7 # [m]
x_dot_0 = 100 # [m/s]
x_dot_0_1 = 10 # [m/s]
x_dot_0_2 = 50 # [m/s]

x = [x_0*np.cos(omega_n*t)+(x_dot_0/omega_n)*np.sin(omega_n*t) for t in t]
x_1 = [x_0*np.cos(omega_n*t)+(x_dot_0_1/omega_n)*np.sin(omega_n*t) for t in t]
x_2 = [x_0*np.cos(omega_n*t)+(x_dot_0_2/omega_n)*np.sin(omega_n*t) for t in t]

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(t, x, c='m',label='$v_{0}=100$ [m/s]')
ax.plot(t, x_1, c='r',label='$v_{0}=10$ [m/s]')
ax.plot(t, x_2, c='orange',label='$v_{0}=50$ [m/s]')
plt.xlabel('tiempo (s)',fontsize=16)
plt.ylabel('$x(t)$ ',fontsize=16)
plt.title('Vibración NO amotiguada con distintas velocidades iniciales', fontsize=16)
ax.legend(loc=1);
```

```python
# Caso sub-amortiguado

t = np.linspace(0,20,1000) # vector tiempo

k  = 1000 # [N/m]
m  = 100 # [kg]
c1 = 90 # [kg/s]
c2 = 20

x_0 = 0.7 # [m]
x_dot_0 = 100 # [m/s]

zeta1 = c1/(2*np.sqrt(k*m))
w0_1   = np.sqrt(k/m)
wd_1   = w0*np.sqrt(1-zeta1**2)
A_1    = np.sqrt(x_0**2+((x_dot_0+zeta1*w0*x_0)/wd)**2)
fid_1  = np.arctan((x_0*wd)/(x_dot_0+zeta1*w0*x_0))

zeta2 = c2/(2*np.sqrt(k*m))
w0_2   = np.sqrt(k/m)
wd_2   = w0*np.sqrt(1-zeta2**2)
A_2    = np.sqrt(x_0**2+((x_dot_0+zeta2*w0*x_0)/wd)**2)
fid_2  = np.arctan((x_0*wd)/(x_dot_0+zeta2*w0*x_0))

x1    = A_1*np.exp(-zeta1*w0_1*t)*np.sin(wd_1*t+fid_1)
x2    = A_2*np.exp(-zeta2*w0_2*t)*np.sin(wd_2*t+fid_2)

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(t, x1, c='r',label='C = 90 [kg/s]')
ax.plot(t, x2, c='orange',label='C = 20 [kg/s]')
plt.xlabel('tiempo (s)',fontsize=16)
plt.ylabel('$x(t)$ ',fontsize=16)
plt.title('Vibración sub-amotiguada con distintos amortiguamientos', fontsize=16)
ax.legend(loc=1);
```

```python
# Caso criticamente amortiguado

t = np.linspace(0,1.5,1000) # vector tiempo

omega_n = 10 # [rad/s]

x_0     = 10 # [m]
x_dot_0 = 200 # [m/s]

x       = np.exp(-omega_n*t)*(x_0+(x_dot_0+omega_n*x_0)*t)

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(t, x, c='r',label='')
plt.xlabel('tiempo [s]',fontsize=16)
plt.ylabel('$x(t) $ [m] ',fontsize=16)
plt.title('Vibración críticamente amortiguada', fontsize=16);
```

```python
# Caso sobre-amortiguado

t = np.linspace(0,10,1000) # vector tiempo

omega_n  = 3 # [rad/s]
zeta   = 4
x_0  = 1 # [m]
v_dot_0 = 9 # [m/s]

x1  = (np.exp(-zeta*omega_n*t)/2*np.sqrt(zeta**2-1))*((v_dot_0/omega_n+x_0*(zeta+np.sqrt(zeta**2-1)))*np.exp(omega_n*t*np.sqrt(zeta**2-1))+(-v_dot_0/omega_n+x_0*(-zeta+np.sqrt(zeta**2-1)))*np.exp(-omega_n*t*np.sqrt(zeta**2-1)))

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(t, x1, c='r',label='')
plt.xlabel('tiempo [s]',fontsize=16)
plt.ylabel('$x(t) $ [m] ',fontsize=16)
plt.title('Vibración sobre-amortiguada', fontsize=16);
```
